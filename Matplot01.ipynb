```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("X-label")
plt.ylabel("y-label")
plt.title("Plot")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.5)
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("random data")
```

```python
x = ['a','b','c','d','e']
```

```python
y = np.random.rand(5)
plt.bar(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar")
```

```python
y 
```

```python
plt.barh(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,'--b')
plt.xlabel("X-label")
plt.ylabel("y-label")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y,c = colours,s = sizes,alpha=.5, marker='v')
```

```python
x = [12,12,1212,1,2,12,12,1]
y=[12,212,2,12,12,2,21,2]
plt.plot(x,y)
plt.show
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
